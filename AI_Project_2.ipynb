Toxicity Detection

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# 1. Chhota sa Dummy Dataset setup (Taaki koi download na karna pade)
data = {
    'comment': [
        'I love this post, it is amazing!',
        'You are an idiot and I hate you',
        'Have a great day ahead my friend',
        'This is worst, shut up and go away'
    ],
    'is_toxic': [0, 1, 0, 1] # 0 = Safe/Good, 1 = Toxic/Bad
}
df = pd.DataFrame(data)

# 2. Vectorizer aur Model Setup
cv = CountVectorizer()
X = cv.fit_transform(df['comment'])
y = df['is_toxic']

# 3. Model Training
model = MultinomialNB()
model.fit(X, y)

# 4. Test Example (Output check karne ke liye)
test_comment = ["You are doing a terrible job"]
test_cv = cv.transform(test_comment)
prediction = model.predict(test_cv)

print("--- Toxicity Detection Output ---")
print(f"Comment: {test_comment[0]}")
print("Result: TOXIC" if prediction[0] == 1 else "Result: SAFE/CLEAN")
```
